```
# read avro datasets
df = spark.read.format("avro").option("header", True).option("inferSchema", True).load("dbfs:/FileStore/shared_uploads/sudippandit99@gmail.com/part_m_00003-1.avro")
```

```
df.show(5)
```

```
#print schema
df.printSchema()
```

```
## add columns
=> syntax is df.withColumn("new_column_name", value)

## rename column
=> syntax is df.withColumnRenamed("Old_name", "New_name")

## Drop column
=> syntax is df.drop('column_name')
```

```
# add column with constant literal
from pyspark.sql.functions import lit
df_new_column = df.withColumn("Country", lit("US")).show(4)
```

```
# add new column by calculations
from pyspark.sql.functions import concat
new_customer_id = df.withColumn("New_customer_id", df.customer_id+10)
new_customer_id.show(5)
# I have just displayed and do dummy calculations on the numerical column as i have but sometimes if we have some columns like salary and we are going to calculate the bonus, at this case we can use the syntax like df.withColumn("column_name", salary*0.001) to calculate the bonus.
# We can add columns as per our business logic and write code as per the logic and add column that we want
```

```
# withColumnRenamed column name
column_renamed = df.withColumnRenamed("customer_id", "ID")
column_renamed.show(5)
```

```
# drop columns
df_drop_columns = new_customer_id.drop("New_customer_id")
df_drop_columns.show(5)
```
