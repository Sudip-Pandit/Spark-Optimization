```
BROADCAST JOIN:
=>It avoids data shuffling in worker nodes, shared variables which are read-only
=> It is cached in each node of the cluster, reduces network I/O operations. 
=> It is best performance tuning techniques. 
=> It is best if one side of join is tiny table(Ideal for joining situations).Only suitable for smaller tables join. 
=> Is is send to the each node of the cluster through driver. 
=> We can define the broadcast join as read-only shared variables
-> through which we can store/cached data into each node of the cluster
-> instead of sending data at each time a task needs it of any operations.

from pyspark.sql import SparkSession
spark = SparkSession.builder \
        .master("local[*]") \
        .appName("Broadcast") \
        .getOrCreate()
```

```
import pyspark.sql.types as st

data = [
    (100, 'Shirts', 1000),
    (200, 'Pants', 2000),
    (300, 'Cosmetics', 3000)
]

user_schema = st.StructType([
    st.StructField('Item_id', st.IntegerType(), True),
    st.StructField('Items', st.StringType(), True),
    st.StructField('Amount', st.IntegerType(), True)
])

df_trans = spark.createDataFrame(data, user_schema)
df_trans.show()
```

```
data1 = [
    (100, 'London'),
    (200, 'US')
]

user_schema = st.StructType([
    st.StructField('Item_id', st.IntegerType(), True),
    st.StructField('Location', st.StringType(), True),
    ])

df_loc = spark.createDataFrame(data1, user_schema)
df_loc.show()
```

```
from pyspark.sql.functions import broadcast
joinDF = df_trans.join(broadcast(df_loc), df_trans['Item_id']==df_loc['Item_id']).show()
```
