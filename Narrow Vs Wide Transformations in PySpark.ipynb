```
### Transformation and Actions:
=> Transformation is an operation technique in spark which would apply to create RDD/Dataframe/Datasets or operation applied to the existing RDD/Dataframe/Datasets.
=> There are two types of transformations 1) Narrow Transformation 2) Wide transformation

Data in clusters are distributed in partitions and each partition is extended to the existing nodes of the cluster. The partition of the cluster is dependent on the cores of the node. So, the large and few number of partitions are not good in spark in terms of optimization techniques. 

Narrow transformation is applied when we call the spark functions such as filter, Map, flatMap, Union etc. While calling these transformations, the operation is independent, it means the spark partition is crystal clear about the data that is going to fetch from each partition, so there is not shuffling of data to produce outcome. It is simple transformation and not expensive. Each and every partitions works independently and return the result to the driver.
=> df2 = df1.filter(df1.employee_id=101)

In wide transformation, data is shuffled across the nodes so it is very expensive transformation and must be ignored as much when we do practice in real time. Examle is groupBy etc. Each partitions are dependent to each others to fetch the data and triggers more and more I/O operations. The process of wide transformation starts from grouping the data from each partition and grouped together having each key, which is only possible having massive shuffling among the nodes. 
=> df2 = df1.groupBy("Department").count()

### Actions
=> In spark, the transformations are called lazy, the DAGs are only updated not execute while calling transformations. The result is produced once when we called actions. When we called actions, the DAG is executed not updated, that's the reason we get out results while calling actions. 
```
