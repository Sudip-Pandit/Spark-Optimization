```
# Spark session is unified entry point of a spark application from spark 2.0
# Spark session is a combination of Spark Context, SQL context and hive context
# Spark context is the entry point to spark programming with RDD and to connect to spark cluster
# Since spark 2.0 spark session has been introduced and became an entry point to start programming and DF and dataset

# SparkContext: Main entry point for spark functionality.
# A SparkContext represents the connection to a spark cluster, and can be used to create RDDs, accumulators and broadcast variables on that cluster.
# SparkSession: the enrty point to programming Spark with the Dataset and dataFrame API.
  
 # Key points:
# 1. Only one SparkContext should be active per JVM. You must stop() the active Sparkcontext before creating a new one.
# 2. We can start new SparkSession with isolated SQL configurations, temporary tables, registered functions, but sharing the underlying SparkContext and
# cache data.
```
